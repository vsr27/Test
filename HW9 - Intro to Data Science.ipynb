```python
observed = results2['TotalScore'].mean() 
expected = results['TotalScore'].mean()

chi_squared_stat = (((observed-expected)**2)/expected)

print(chi_squared_stat)
```

```python
crit = stats.chi2.ppf(q = 0.95, # Find the critical value for 95% confidence*
                      df = 2)   # Df = number of variable categories - 1

print("Critical value")
print(crit)
# Find the p-value
p_value = 1 - stats.chi2.cdf(x=chi_squared_stat,  
                             df=1)
print("P value")
print(p_value)
```

Since our chi-squared statistic less the critical value, we'd accept the null hypothesis that the two distributions are the same.
