## Step 2: Quick Data Exploration

```python
#Renaming the column names as preferred for further use

column_names = {'DBN': 'DBN',  'SCHOOL NAME':'School_Name',
                 'Num of SAT Test Takers': 'No_of_SAT_Test_Takers', 'SAT Critical Reading Avg. Score':'SAT_Critical_Reading_Avg_Score',
                'SAT Math Avg. Score':'SAT_Math_Avg_Score',
                'SAT Writing Avg. Score':'SAT_Writing_Avg_Score'}   

results = (results.rename(columns=column_names)
    .dropna(thresh=4)
    [['DBN', 'School_Name', 'No_of_SAT_Test_Takers', 'SAT_Critical_Reading_Avg_Score', 'SAT_Math_Avg_Score','SAT_Writing_Avg_Score']]
        .set_index('DBN', append=True)
     .rename_axis(["Sr No", "DBN"])
    .sort_index())

#rename() : function takes as argument the dict of new column names
#dropna() : Returns object with labels on given axis omitted where alternately any or all of the data are missing
#assign() : function helps to change the date format using to_datetime function of pandas
#set_index() :Sets the DataFrame index (row labels) using one or more existing columns and by default yields a new object.
#rename_axis() :Alters index and / or columns using input function or functions.A scalar or list-like for mapper will alter the Index.name or MultiIndex.names attribute.
#sort_index() : Sorts object by labels (along an axis)
 
results.head(600) 
```

### b) Describe a second way to answer the same question?

We can expect that schools having a greater number of students taking the exam will hold the scores at a certain range: with a smaller sample size we would expect scores having more variance.
We could also find the difference between the top and bottom 10% of schools. We could see whether we can support the hypothesis that high schools where a greater number of students take the SAT are more academically rigorous or better-equipped.

Coming to Others Scores, one intuition can be that some test takers have lower scores in one section but strong in others.
So it might be a good idea to combine all scores and take a log transformation of the same.

```python
results['TotalScore'] = results['SAT_Critical_Reading_Avg_Score'] + results['SAT_Math_Avg_Score']+results['SAT_Writing_Avg_Score']
results['TotalScore_log'] = np.log(results['TotalScore'])
results['TotalScore_log'].hist(bins=20) 

#Now the distribution looks much closer to normal and effect of extreme values has been significantly subsided.
```
